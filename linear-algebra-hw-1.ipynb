   $\sqrt{1/2+1/2+0}=1$<br>
   $\sqrt{1/2+1/2+0}=1$<br>
   $\sqrt{0+0+1}=1$<br>

   $\sqrt{1/4+1/4+0}=\sqrt {1/2}$<br>
   $\sqrt{0+1/4+1/4}=\sqrt {1/2}$<br>
   $\sqrt{0+0+1}=1$<br>
   
г) $(1,0,0),(0,1,0),(0,0,1)$ - ортонормированный базис
