# Задание №1

Найдите собственные векторы и собственные значения для линейного оператора, заданного матрицей:

$$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}.$$

## Решение:

Найдем собственные значения матрицы $A$:

$\begin{pmatrix}
-1 -\lambda & -6\\
2 & 6 - \lambda
\end{pmatrix}=0, \quad
(-1-\lambda)(6-\lambda)-2\cdot(-6)=0, \quad
\lambda^2-5\lambda+6=0, \quad
\lambda=(2,3)$

Для каждого собственного значения найдем собственный вектор:
$Ax=\lambda x, \quad\
(A-\lambda E)x=0$
$\lambda = 2:$
$\begin{pmatrix}
-1 -2 & -6\\
2 & 6 - 2\
\end{pmatrix}\cdot \begin{pmatrix}
x_1\\
x_2\\
\end{pmatrix}=0, \quad
\begin{cases}
-3x_1 -6x_2 = 0\\
2x_1 +4x_2 = 0\end{cases}, \quad
\begin{cases}
x_1 =-2x_2\\
x_1 =-2x_2\end{cases}, \quad
x=(2, -1)$
    
Проверим, что найденный собственный вектор удовлетворяет каждому уравнению:
$\begin{cases}
-3\cdot 2 -6\cdot(-1) = 0\\
2\cdot2 +4\cdot(-1) = 0\end{cases}$
    
$\lambda = 3:$
$\begin{pmatrix}
-1 -3 & -6\\
2 & 6 - 3\
\end{pmatrix}\cdot \begin{pmatrix}
x_1\\
x_2\\
\end{pmatrix}=0, \quad
\begin{cases}
-4x_1 -6x_2 = 0\\
2x_1 +3x_2 = 0\end{cases}, \quad,
\begin{cases},
x_1 =-3/2x_2\\
x_1 =-3/2x_2\end{cases}, \quad
x=(3, -2)$

Проверим, что найденный собственный вектор удовлетворяет каждому уравнению:
$\begin{cases}
-4\cdot3 -6\cdot(-2) = 0\\
2\cdot3 +3\cdot(-2) = 0\end{cases}$

Ответ:

Собственные значения матрицы: $\lambda=(2,3)$
Собственные вектора матрицы: $x=(2,-1), x=(3,-2)$

# Задание №3

Пусть линейный оператор задан матрицей:

$$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$$

Установите, считается ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.

## Решение:

$\begin{pmatrix}
1 & 1\\
-1 & 3
\end{pmatrix}\cdot
\begin{pmatrix}
1\\
1\end{pmatrix}=\lambda\cdot\begin{pmatrix}
1\\
1\end{pmatrix}.$
    
$\begin{cases}
1 + 1 = \lambda\\
-1 + 3 = \lambda
\end{cases}, \quad
\begin{cases}
\lambda = 2\\
\lambda = 2
\end{cases}$
Ответ:

Вектор $x=(1,1)$ является собственным вектором этого линейного оператора

# Задание №4

Пусть линейный оператор задан матрицей:

$$A=\begin{pmatrix} 0 & 3 & 0\\ 3 & 0 & 0\\ 0 & 0 & 3 \end{pmatrix}.$$

Установите, считается ли вектор $x=(3, -3, -4)$ собственным вектором этого линейного оператора.

## Решение:

$\begin{pmatrix}
0 & 3 & 0\\
3 & 0 & 0\\
0 & 0 & 3 \end{pmatrix}\cdot
\begin{pmatrix}
3\\
-3\\
-4\end{pmatrix}=\lambda \cdot \begin{pmatrix}
3\\
-3\\
-4\end{pmatrix}.$
    
$\begin{cases}
3\cdot(-3)=3\lambda\\
3\cdot3=-3\lambda\\
3\cdot(-4)=-4\lambda \end{cases}, \quad
\begin{cases}
\lambda=-3\\
\lambda=-3\\
\lambda=3 \end{cases}$
Ответ:
Вектор $x=(3, -3, -4)$ не является собственным вектором этого линейного оператора.
